Starting with the cast feature

From the above we can see that for some of the movies there are about 50 directors and thats why there are about 50 columns got created.

```python
df.loc[df['cast'].str.split(',',expand=True)[49].isin([' Lovie Simone', ' Rafe Spall'])]['cast']
```

The above 2 are the movies that have about 50 directors. Thus we have to separate out these cast members into separate records and continue the analysis further.

```python
data = df.copy(deep=True)   #Creating restore point for the dataset.
```

We can see there are comma separated values in the cast records. Thus we will split them based on ',' (comma) separator and then convert them into list and store them in the same place.

Using lambda function and .apply() method, we can convert the string data into list of comma separated values.

```python
df['cast'] = df['cast'].apply(lambda x : list(x.split(',')) if type(x)==str else x)
df
```

```python
data = df.copy(deep=True)   #Creating restore point for the dataset.
```

##### Example of cast details for one of the TV show.

With the help of .explode() method we can explode the list of cast details into individual records which thus creates a unique director, actor pair.

```python
exp_df = df.explode('cast')
exp_df
```

We can see the number of records have increased in the exp_df when compared to the original dataset.

```python
df.loc[df['show_id']=='s2']['cast'].values
```

```python
exp_df.loc[exp_df['show_id']=='s2']
```

Similar to Cast feature of the dataframe, there are other features like Director, Country and listed_in which has to be un-nested in the dataset.

We will apply the same split and explode method for these features also.

```python
exp_df['director'] = exp_df['director'].apply(lambda x:list(x.split(', ')) if type(x)==str else x)   #Splitting the string data into list of strings.
exp_df['country'] = exp_df['country'].apply(lambda x:list(x.split(', ')) if type(x)==str else x)
exp_df['listed_in'] = exp_df['listed_in'].apply(lambda x:list(x.split(', ')) if type(x)==str else x)
```

##### Exploding the features one by one:

```python
exp_df = exp_df.explode("director")
exp_df = exp_df.explode("country")
exp_df = exp_df.explode("listed_in")
exp_df.reset_index(drop=True,inplace=True)
exp_df
```
